```python
al = [chr(i) for i in range(97, 123)]
nu = [i for i in range(1, 27)]

pairs = dict(zip(nu, al))
pairs[0] = 'z'
```

```python
def encode(text, key):
    new_str = ''
    for i in text:
        
        if i.isalpha():
            if i.isupper():
                new_str += pairs[(ord(i)-64+key)%26].upper()
            else:
                new_str += pairs[(ord(i)-96+key)%26]
            
        else:
            new_str += i
            
    return new_str
            

def decode(text, key):
    new_str = ''
    
    for i in text:
        
        if i.isalpha():
            if i.isupper():
                new_str += pairs[(ord(i)-64-key)%26].upper()
            else:
                new_str += pairs[(ord(i)-96-key)%26]
            
        else:
            new_str += i
            
    return new_str

def decode_advance(text):
    
    possible_answers = []
    
    one = ['a', 'i']
    
    two = [ 'of', 'to', 'in', 'it', 'is', 'be', 
           'as', 'at', 'so', 'we', 'he', 'by', 'or', 
           'on', 'do', 'if', 'me', 'my', 'up', 'an', 'go', 'no', 'us', 'am', 'hi']
    
    three = ['the', 'and', 'for', 'are', 'but', 'not', 'you', 'all', 'any', 
             'can', 'had', 'her', 'was', 'one', 'our', 'out', 'day', 'get', 
             'has', 'him', 'his', 'how', 'man', 'new', 'now', 'old', 'see', 
             'two', 'way', 'who', 'boy', 'did', 'its', 'let', 'put', 'say', 
             'she', 'too', 'use', 'hey']
    
    four = ['that', 'with', 'have', 'this', 'will', 'your', 'from', 'they', 
            'know', 'want', 'been', 'good', 'much', 'some', 'time', 'hello']
    
#     common_words = [two, three, four]
    
    mylist = text.split()
    
    for i in mylist:
        
        if(len(i) == 1):
            for j in range(26):
                if decode(i, j).lower() in one:
                       possible_answers.append(decode(text, j))
        if(len(i) == 2):
            for j in range(26):
                if decode(i, j).lower() in two:
                    possible_answers.append(decode(text, j))
        if(len(i) == 3):
            for j in range(26):
                if decode(i, j).lower() in three:
                    possible_answers.append(decode(text, j))
        if(len(i) == 4):
            for j in range(26):
                if decode(i, j).lower() in four:
                    possible_answers.append(decode(text, j))
     
    max_ans_freq = 0 
    ans = ''
    for i in possible_answers:
        if possible_answers.count(i) > max_ans_freq:
            max_ans_freq = possible_answers.count(i)
            ans = i
            if(max_ans_freq > int(len(possible_answers)/2)):
                return i
    
    return i
```
